```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
df=pd.concat([train,test],sort=False)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Fare']=df['Fare'].fillna(df['Fare'].median())
df['Cabin']=df['Cabin'].fillna('Missing')
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked']=df['Embarked'].fillna('S')
```

```python
df.isnull().sum()
```

```python
df['Cabin']=df['Cabin'].str[0]
```

```python
df['Cabin'].value_counts()
```

```python
df.head()
```

```python
df['SibSp'].value_counts()
```

```python
import re
def get_title(name):
    title_search = re.search(' ([A-Za-z]+\.)', name)
    
    if title_search:
        return title_search.group(1)
    return ""
```

```python
df['Title'] = df['Name'].apply(get_title)
df['Title'].value_counts()
```

```python
df['Title'] = df['Title'].replace(['Capt.', 'Dr.', 'Major.', 'Rev.'], 'Officer.')
df['Title'] = df['Title'].replace(['Lady.', 'Countess.', 'Don.', 'Sir.', 'Jonkheer.', 'Dona.'], 'Royal.')
df['Title'] = df['Title'].replace(['Mlle.', 'Ms.'], 'Miss.')
df['Title'] = df['Title'].replace(['Mme.'], 'Mrs.')
df['Title'].value_counts()
```

```python
df.loc[df['Age'] <= 16, 'Age'] = 0
df.loc[(df['Age'] > 16) & (df['Age'] <= 32), 'Age'] = 1
df.loc[(df['Age'] > 32) & (df['Age'] <= 48), 'Age'] = 2
df.loc[(df['Age'] > 48) & (df['Age'] <= 64), 'Age'] = 3
df.loc[ df['Age'] > 64, 'Age'] = 4 
```

```python
df.head()
```

```python
df.head()
```

```python
df_eve=df.drop(['Name','Ticket'],axis=1)
```

```python
df_eve.head()
```

```python
df_dummies=pd.get_dummies(df_eve,drop_first=True)
```

```python
df_dummies.columns
```

```python
df_dummies.head()
```

```python
df_train=df_dummies[df_dummies['Survived'].notna()]
```

```python
df_test=df_dummies[df_dummies['Survived'].isna()]
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train.drop(['PassengerId','Survived'],axis=1), 
                                                    df_train['Survived'], test_size=0.30, 
                                                    random_state=101, stratify = df_train['Survived'])
```

```python
from sklearn.linear_model import LogisticRegression
Logreg=LogisticRegression()
Logreg.fit(X_train,y_train)
```

```python
y_pred=Logreg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
TestForPred = df_test.drop(['PassengerId', 'Survived'], axis = 1)
```

```python
t_pred = Logreg.predict(TestForPred).astype(int)
```
